```python
font1 = {'family':'serif','color':'blue','size':20}
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1,y1,x2,y2) 
plt.title("Line graph" ,fontdict=font1, loc='left') # loc - left, right, center
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(axis='x') # axis - x, y, both
plt.grid(axis='y')

plt.grid(color = 'green', linestyle = ':', linewidth = .5)

plt.show()
```

```python
plt.suptitle("Graph " )
plt.title("Graph 1" ) 
plt.plot(x1,y1,"r")
plt.show()
plt.title("Graph 2" ) 
plt.plot(x2,y2,"g")
plt.show()
```

```python
#Scatter
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.subplot(1,2,1)
plt.scatter(x,y,c="red")
plt.subplot(1,2,2)
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
# BarGraph
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.subplot(1,4,1)
plt.bar(x,y)
plt.subplot(1,4,2)
plt.barh(x,y)
plt.subplot(1,4,3)
plt.bar(x,y,color = "red")
plt.show()
```

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
y= np.array(["35","20","49","20"])
plt.pie(y)
label = ["A","B","C","D"]
plt.pie(y , labels = label , startangle = 0)
plt.show()
```
