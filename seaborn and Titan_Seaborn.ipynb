```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("titan.csv")
```

Histplot = for histograms with optional density; add kde=True for a smoothed estimate
Kdeplot = for continuos variables .

```python
sns.histplot(data = df , x='Age', bins=30, kde=True)
plt.title("Age distribution with kde")
plt.show()
```

```python
sns.kdeplot(data = df , x='Age', fill=True,common_norm = False)
plt.title("Fare Distribution")
plt.show()
```

#### countplot = shows category frequencies
#### barplot = shows an aggregate and confidence interval i.e here mean survived by Pclass
#### boxplot/violinplot reveal distribution shape across groups.

```python
sns.countplot(data=df, x='Sex' , hue = "Survived")
plt.title("Survival by gender")
plt.show()
```

```python
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title("Average Survival rate  by Passenger Class")
plt.show()
```

```python
sns.boxplot(data = df , x = "Pclass" , y = "Fare")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
sns.violinplot(data=df, x='Sex', y='Age')
plt.title("Age Distribution by Sex")
plt.show()
```

#### sctterplot = takes the data  with parameters x and y with hue(representing True or False) and style for differentiating various group
                    and alpha determines the brightness of the style
#### lineplot = The line shows the average survival rate (mean of Survived) at each age.

                The shaded area (if shown) indicates the confidence interval around the mean estimate.

                By sorting Age, the line progresses smoothly from youngest to oldest.

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', style='Pclass', alpha=1)
plt.title("Age vs Fare vs Survival")
plt.show()
```

```python
sns.lineplot(data=df.sort_values('Age'), x='Age', y='Survived', estimator='mean', n_boot=1000)
plt.title("Survival trend by Age")
plt.show()
```

```python
corr = df[['Age','Fare','SibSp','Parch','Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue='Survived')
plt.show()
```
